### **What is an A/B test?**
A/B tests are experiments that compare two scenarios (e.g., scenario A vs. scenario B).
They test subjects' responses to each of the variants to determine which is more effective ([read more about A/B tests on Wikipedia](https://en.wikipedia.org/wiki/A/B_testing)).  
A/B tests allow us to draw conclusions about the different scenarios by randomizing exposure to them, e.g., one could flip a coin and assign a user to scenario A if it lands heads and to B if it lands tails.  
Since exposure is randomized, we can be confident that the scenarios are the sole explanation for statistically significant differences in subjects' responses (if they exist).  
In theory, A/B testing refers to an experiment that compares two scenarios; however, in practice, the term is also used when we compare multiple scenarios (e.g., A vs. B vs. C), although the more precise terminology would be to call such an experiment a "multinomial test."

### **How were A/B tests used by Upworthy?**
Upworthy used A/B testing to increase news readership, conducting experiments for each published news piece.  
In each experiment, they created multiple "packages" of stimuli, varying headlines, images, excerpts, and **ledes** for the same news piece.  
Different "packages" were shown on their (now defunct) website to engage users with the news pieces they produced. Upworthy found "the best" package by conducting A/B tests, showing different packages to different users, and measuring how often users clicked on each version.  
Below, we show three "packages" used by Upworthy in an experiment, each with a different headline for the same news piece.  
Upworthy randomized users that visited their website saw one of the three versions of the headline below. Then, they measured the percentage of times users in each scenario clicked to read the news.  
The headline with the highest percentage of clicks per view (click through rate) was then declared the "winner" and became the default for all visitors.

1.7 For our subsequent analysis, we want to compare the causal effect of headlines on the success of a news piece.
For that, we can compare pairs of packages with the same `eyecatcher_id`, `lede`, and `excerpt`, but different `headlines`.
Note that this means that if an experiment considered 5 different headlines and did not vary any other stimulus, we would have 5C2 (i.e., 5 choose 2, 10) pairs to consider.
Create a dataset where:
- Each row corresponds to a pair of packages with different `headline` but the same `eyecatcher_id`, `lede`, and `excerpt`.
- There are columns containing the headlines of each of the news versions (`headline1`, `headline2`) and the clickthrough rate of each of the news versions (`ctr1`, `ctr2`).
- The columns `headline1` and `ctr1` contain the data associated with the news version with the highest clickthrough rate. Print the first rows of your newly created dataframe, as well as its length.
-  The rows where the two news pieces had exactly the same clickthrough rate should be filtered out (this is for simplicity's sake).
-  The column `date_created` contains the date when the news version with the highest clickthrough rate was created.


```python
from itertools import combinations

# Create a function to generate pairs of packages with different headlines
def create_headline_pairs(df):
    pairs_list = []

    # Grouping by eyecatcher_id, lede, and excerpt
    grouped = df.groupby(['eyecatcher_id', 'lede', 'excerpt'])

    # Iterate through each group
    for _, group in grouped:
        headlines = group['headline'].tolist()

        # Generate pairs of headlines within the same group
        headline_combinations = combinations(headlines, 2)

        for pair in headline_combinations:
            # Calculate CTR for each headline
            ctr1 = group[group['headline'] == pair[0]]['ctr'].mean()
            ctr2 = group[group['headline'] == pair[1]]['ctr'].mean()

            # Filter out rows where both headlines have the same CTR
            if ctr1 != ctr2:
                date_created_1 = group[group['headline'] == pair[0]]['created_at'].iloc[0]
                date_created_2 = group[group['headline'] == pair[1]]['created_at'].iloc[0]

                # Select the higher CTR as headline1 and its associated data
                if ctr1 > ctr2:
                    pairs_list.append([pair[0], pair[1], ctr1, ctr2, date_created_1])
                else:
                    pairs_list.append([pair[1], pair[0], ctr2, ctr1, date_created_2])

    # Create a DataFrame from the pairs_list
    pairs_df = pd.DataFrame(pairs_list, columns=['headline1', 'headline2', 'ctr1', 'ctr2', 'date_created'])

    return pairs_df

# Create the dataset containing pairs of packages with different headlines
pairs_df = create_headline_pairs(df)

# Filter out rows where both news pieces had the same clickthrough rate
pairs_df = pairs_df[pairs_df['ctr1'] != pairs_df['ctr2']]

# Display the first rows and length of the newly created dataframe
print("\nLength of the dataframe:", len(pairs_df))
print("First rows of the newly created dataframe:")
pairs_df.head()
```

```python
# Calculate the average difference between ctr1 and ctr2
average_difference = (pairs_df['ctr1'] - pairs_df['ctr2']).mean()

# Calculate the average clickthrough rate for ctr2
average_ctr2 = pairs_df['ctr2'].mean()

print("Average difference between most and least clicked-through headlines:", average_difference)
print("Average clickthrough rate for the least clicked-through headline (ctr2):", average_ctr2)
```

Each indicator variable in the dataframe should equal 1 if the corresponding headline uses the corresponding type of pronoun and 0 otherwise.

```python
# Function to create indicator variables for each set of pronouns
def create_pronoun_indicators(headline, pronoun_list):
    # Convert headline to lowercase for case-insensitive comparison
    headline_lower = headline.lower()
    # Check if any word in the headline matches the pronouns in the given pronoun list
    return int(any(word in headline_lower.split() for word in pronoun_list))

# Create indicator variables for each set of pronouns in the DataFrame
for feature, pronouns in feature_wordsets.items():
    pairs_df[f'{feature}1'] = pairs_df['headline1'].apply(lambda x: create_pronoun_indicators(x, pronouns))
    pairs_df[f'{feature}2'] = pairs_df['headline2'].apply(lambda x: create_pronoun_indicators(x, pronouns))
```

```python
# Function to read positive or negative words from file and extract the words
def read_sentiment_words(file_path):
    sentiment_words = set()

    # Try reading the file with different encodings until successful or an error occurs
    possible_encodings = ['utf-8', 'latin-1', 'utf-16', 'cp1252']
    for encoding in possible_encodings:
        try:
            with open(file_path, 'r', encoding=encoding) as file:
                lines = file.readlines()
                found_start = False

                # Extract words from the word list
                for line in lines:
                    if found_start:
                        words = line.strip().lower().split()
                        sentiment_words.update(words)
                    elif line.strip().startswith(';'):
                        found_start = True

            break  # Break loop if successful reading
        except UnicodeDecodeError:
            pass  # Try the next encoding if decoding error occurs

    return sentiment_words

# File paths for positive and negative sentiment word lists
positive_file_path = 'data/positive-words.txt'
negative_file_path = 'data/negative-words.txt'

# Read positive and negative sentiment words
positive_words = read_sentiment_words(positive_file_path)
negative_words = read_sentiment_words(negative_file_path)

# Function to check if a headline contains positive sentiment words
def has_positive_sentiment(headline, positive_words):
    words = set(headline.lower().split())
    return int(bool(words.intersection(positive_words)))

# Function to check if a headline contains negative sentiment words
def has_negative_sentiment(headline, negative_words):
    words = set(headline.lower().split())
    return int(bool(words.intersection(negative_words)))

# Create indicator columns for positive and negative sentiment in headline1 and headline2
pairs_df['positive1'] = pairs_df['headline1'].apply(lambda x: has_positive_sentiment(x, positive_words))
pairs_df['positive2'] = pairs_df['headline2'].apply(lambda x: has_positive_sentiment(x, positive_words))
pairs_df['negative1'] = pairs_df['headline1'].apply(lambda x: has_negative_sentiment(x, negative_words))
pairs_df['negative2'] = pairs_df['headline2'].apply(lambda x: has_negative_sentiment(x, negative_words))
```

The expected values and variances for the generated random variables are:

- Expected Value ($E[X]$): 0.5
- Variance ($Var(X)$): $\frac{1}{12} \approx 0.0833$

For $Y \sim Uniform(0,1)$:
- Expected Value ($E[Y]$): 0.5
- Variance ($Var(Y)$): $\frac{1}{12} \approx 0.0833$

For $Z = \frac{X}{2} + \frac{Y}{2} + 0.1$:
- Expected Value ($E[Z]$): $E[\frac{X}{2} + \frac{Y}{2} + 0.1] = \frac{E[X]}{2} + \frac{E[Y]}{2} + 0.1 = 0.5 \times \frac{1}{2} + 0.5 \times \frac{1}{2} + 0.1 = 0.6$
- Variance ($Var(Z)$): $Var(\frac{X}{2} + \frac{Y}{2} + 0.1) = Var(\frac{X}{2}) + Var(\frac{Y}{2}) = \frac{1}{4} \times Var(X) + \frac{1}{4} \times Var(Y) = 0.0833 \times \frac{1}{4} + 0.0833 \times \frac{1}{4} = 0.0417$

For $K = Y + 0.1$:
- Expected Value ($E[K]$): $E[Y + 0.1] = E[Y] + 0.1 = 0.5 + 0.1 = 0.6$
- Variance ($Var(K)$): $Var(Y + 0.1) = Var(Y) = 0.0833$

Using python:

```python
# Calculate expected values and variances
expected_X = np.mean(X)
expected_Y = np.mean(Y)
expected_Z = np.mean(Z)
expected_K = np.mean(K)

var_X = np.var(X)
var_Y = np.var(Y)
var_Z = np.var(Z)
var_K = np.var(K)

# Display results
print(f"Expected Value of X: {expected_X:.4f}, Variance of X: {var_X:.4f}")
print(f"Expected Value of Y: {expected_Y:.4f}, Variance of Y: {var_Y:.4f}")
print(f"Expected Value of Z: {expected_Z:.4f}, Variance of Z: {var_Z:.4f}")
print(f"Expected Value of K: {expected_K:.4f}, Variance of K: {var_K:.4f}")
```

3.5 Run the following simulation 10000 times, storing the $p$-values for the tests at each run:
- Sample new values  for $X$, $Y$, $Z$ and $K$ ($n=30$ each).
- Run independent sample t-test (assuming equal variance) and paired t-test comparing $X$ and $Z$.
-  Run independent sample t-test (assuming equal variance) and paired t-test comparing $X$ and $K$.

```python
np.random.seed(42)

# Initialize arrays to store p-values
p_values_X_Z_ind = []
p_values_X_Z_pair = []
p_values_X_K_ind = []
p_values_X_K_pair = []

# Run simulation 10000 times
for _ in range(10000):
    # Generate new samples for X, Y, Z, and K
    X = np.random.uniform(0, 1, 30)
    Y = np.random.uniform(0, 1, 30)
    Z = X/2 + Y/2 + 0.1
    K = Y + 0.1

    # Perform t-tests and store p-values
    p_values_X_Z_ind.append(stats.ttest_ind(X, Z).pvalue)
    p_values_X_Z_pair.append(stats.ttest_rel(X, Z).pvalue)
    p_values_X_K_ind.append(stats.ttest_ind(X, K).pvalue)
    p_values_X_K_pair.append(stats.ttest_rel(X, K).pvalue)

# Display some of the p-values (for illustration)
print("Sample p-values for X vs Z (Independent t-test):", p_values_X_Z_ind[:5])
print("Sample p-values for X vs Z (Paired t-test):", p_values_X_Z_pair[:5])
print("Sample p-values for X vs K (Independent t-test):", p_values_X_K_ind[:5])
print("Sample p-values for X vs K (Paired t-test):", p_values_X_K_pair[:5])
```

3.6 Recall that the power of a binary hypothesis test is the probability that the test correctly rejects the null hypothesis when the alternative hypothesis is true. Using the p-values and assuming that we reject the null hypothesis if $p < 0.05$, calculate the statistical power of:
- The independent sample t-test comparing $X$ and $Z$.
- The paired t-test comparing $X$ and $Z$.
- The independent sample t-test comparing $X$ and $K$.
- The paired t-test comparing $X$ and $K$.

```python
# Assuming significance level
alpha = 0.05

# Calculate statistical power based on p-values (assuming null hypothesis is false)
power_X_Z_ind = np.mean(np.array(p_values_X_Z_ind) < alpha)
power_X_Z_pair = np.mean(np.array(p_values_X_Z_pair) < alpha)
power_X_K_ind = np.mean(np.array(p_values_X_K_ind) < alpha)
power_X_K_pair = np.mean(np.array(p_values_X_K_pair) < alpha)

# Display results
print(f"Statistical power (Independent t-test X vs Z): {power_X_Z_ind:.4f}")
print(f"Statistical power (Paired t-test X vs Z): {power_X_Z_pair:.4f}")
print(f"Statistical power (Independent t-test X vs K): {power_X_K_ind:.4f}")
print(f"Statistical power (Paired t-test X vs K): {power_X_K_pair:.4f}")
```

3.7 **Discuss:** When are paired t-tests helpful? Justify.

Paired t-tests are particularly useful in scenarios where you have paired or related observations within two groups. Here are some situations where paired t-tests are helpful and justified:

1. **Before-and-after Studies:** Paired t-tests are valuable for assessing changes within the same subjects over time. For instance, in medical studies evaluating the effectiveness of a treatment, researchers often measure the same individuals before and after treatment. Paired t-tests can determine if there's a significant difference in outcomes pre- and post-treatment.

2. **Matched Samples:** In experiments where subjects are paired or matched based on similar characteristics (age, gender, etc.), paired t-tests help assess the impact of a treatment or intervention by comparing outcomes within these matched pairs.

3. **Testing Repeated Measures:** When measurements are taken at different time points or under different conditions on the same subjects, paired t-tests account for individual variability by analyzing the differences between paired observations.

4. **Reducing Variability:** By comparing differences within paired data points, paired t-tests effectively reduce the impact of variability among individuals. This makes them more sensitive in detecting true effects because they focus on changes within the paired observations.

5. **Efficiency in Data Collection:** Paired designs often require fewer participants than independent designs to achieve the same statistical power. This efficiency makes them advantageous when resources, time, or availability of participants are limited.

6. **Controlling for Confounding Factors:** Paired t-tests control for individual differences or other sources of variability that remain constant across the paired observations, thus reducing the influence of potential confounding factors on the results.

In summary, paired t-tests are helpful when dealing with related data points, whether they involve before-and-after measurements, matched pairs, or repeated measures. They offer greater sensitivity in detecting differences within paired observations, making them a valuable tool in various scientific, medical, and social research scenarios.
